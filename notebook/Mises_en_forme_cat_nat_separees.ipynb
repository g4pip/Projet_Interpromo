```python
# Extraction du d√©partement (en supprimant les 3 derniers chiffres du code INSEE)
df_georisques["D√©partement"] = df_georisques["Code INSEE"].astype(str).str[:-3]
```

```python
# Extraction des mois et ann√©es en pr√©cisant que le jour est en premier (format DD/MM/YYYY)
df_georisques["Mois_debut"] = pd.to_datetime(df_georisques["date_debut_evt"], format="%d/%m/%Y", errors='coerce').dt.month
df_georisques["Ann√©e_debut"] = pd.to_datetime(df_georisques["date_debut_evt"], format="%d/%m/%Y", errors='coerce').dt.year

df_georisques["Mois_fin"] = pd.to_datetime(df_georisques["date_fin_evt"], format="%d/%m/%Y", errors='coerce').dt.month
df_georisques["Ann√©e_fin"] = pd.to_datetime(df_georisques["date_fin_evt"], format="%d/%m/%Y", errors='coerce').dt.year
```

```python
# S√©lection des colonnes finales
df_georisques = df_georisques[["Code INSEE", "D√©partement", "Mois_debut", "Ann√©e_debut", "Mois_fin", "Ann√©e_fin", "Type catastrophe"]]

# Affichage du DataFrame transform√©
df_georisques.head()
```

```python
# Dictionnaire de correspondance entre les types de catastrophes actuels et les nouveaux types
mapping_catastrophes = {
    'Avalanche': 'Neige/Gr√™le',
    'Poids de la neige': 'Neige/Gr√™le',
    'Gr√™le': 'Neige/Gr√™le',
    'Coul√©es de boue': 'Inondation',
    'Effondrement et/ou Affaisement': 'Mouvement de terrain',
    'Inondations et/ou Coul√©es de Boue': 'Inondation',
    'S√©cheresse': 'S√©cheresse',
    'Mouvement de Terrain': 'Mouvement de terrain',
    'Inondations Remont√©e Nappe': 'Inondation',
    'Glissement de Terrain': 'Mouvement de terrain',
    'Lave Torrentielle': 'Inondation',
    'Glissement et Effondrement de Terrain': 'Mouvement de terrain'
}
```

```python
# Remplacement des types de catastrophes selon le mapping
df_georisques["Type catastrophe"] = df_georisques["Type catastrophe"].map(mapping_catastrophes)

# Suppression des lignes o√π "Type catastrophe" est NaN (c'est-√†-dire celles qui n'√©taient pas dans le mapping)
df_georisques = df_georisques.dropna(subset=["Type catastrophe"])
```

### Fichier catastrophes par mois

```python
# Initialisation des compteurs
lignes_traitees = 0
lignes_ajoutees = 0

# Fonction pour exploser les √©v√©nements sur plusieurs mois si n√©cessaire
def expand_event(row):
    global lignes_traitees, lignes_ajoutees  # Acc√®s aux compteurs globaux
    lignes_traitees += 1  # Incr√©mente le compteur de lignes trait√©es

    mois_debut = row["Mois_debut"]
    annee_debut = row["Ann√©e_debut"]
    mois_fin = row["Mois_fin"]
    annee_fin = row["Ann√©e_fin"]

    # Liste pour stocker les nouvelles lignes
    expanded_rows = []

    # Boucle pour g√©n√©rer chaque mois entre Mois_debut/Ann√©e_debut et Mois_fin/Ann√©e_fin
    while (annee_debut < annee_fin) or (mois_debut <= mois_fin):
        #print(f"Ann√©e mois d√©but: {annee_debut}, {mois_debut} / Ann√©e mois fin: {annee_fin}, {mois_fin}")
        nouvelle_ligne = {
            "Code INSEE": row["Code INSEE"],
            "D√©partement": row["D√©partement"],
            "Mois": mois_debut,
            "Ann√©e": annee_debut,
            "Type catastrophe": row["Type catastrophe"]
        }
        expanded_rows.append(nouvelle_ligne)

        lignes_ajoutees += 1  # Incr√©mente le compteur de nouvelles lignes

        # Affichage dynamique du compteur
        print(f'\rLignes trait√©es : {lignes_traitees} | Lignes ajout√©es : {lignes_ajoutees}', end='', flush=True)
        # Incr√©menter le mois
        mois_debut += 1

        # Si on a atteint la date de fin, on arr√™te
        if (annee_debut == annee_fin and mois_debut > mois_fin):
            return expanded_rows

        # G√©rer le changement d'ann√©e
        if mois_debut > 12:
            mois_debut = 1
            annee_debut += 1

    return expanded_rows
```

```python
# Appliquer la fonction √† chaque ligne et aplatir les r√©sultats
df_exploded_mois = df_georisques.apply(expand_event, axis=1).explode().apply(pd.Series)
```

### Fichier catastrophes par ann√©e

```python
# Initialisation des compteurs
lignes_traitees = 0
lignes_ajoutees = 0

# Fonction pour exploser les √©v√©nements sur plusieurs ann√©es si n√©cessaire
def expand_event_by_year(row):
    global lignes_traitees, lignes_ajoutees  # Acc√®s aux compteurs globaux
    lignes_traitees += 1  # Incr√©mente le compteur de lignes trait√©es

    annee_debut = row["Ann√©e_debut"]
    annee_fin = row["Ann√©e_fin"]

    # Liste pour stocker les nouvelles lignes
    expanded_rows = []

    # Boucle pour g√©n√©rer chaque ann√©e entre Ann√©e_debut et Ann√©e_fin
    for annee in range(annee_debut, annee_fin + 1):  # Inclut l'ann√©e de fin
        nouvelle_ligne = {
            "Code INSEE": row["Code INSEE"],
            "D√©partement": row["D√©partement"],
            "Ann√©e": annee,  # Seule l'ann√©e est conserv√©e
            "Type catastrophe": row["Type catastrophe"]
        }
        expanded_rows.append(nouvelle_ligne)

        lignes_ajoutees += 1  # Incr√©mente le compteur de nouvelles lignes

        # Affichage dynamique du compteur
        print(f'\rLignes trait√©es : {lignes_traitees} | Lignes ajout√©es : {lignes_ajoutees}', end='', flush=True)

    return expanded_rows
```

```python
# Appliquer la fonction √† chaque ligne et aplatir les r√©sultats
df_exploded_annee = df_georisques.apply(expand_event_by_year, axis=1).explode().apply(pd.Series)
```

```python
# Exporter le DataFrame en CSV
df_exploded_annee.to_csv(path_clean+"catastrophes_nat_georisques_annuel.csv", index=False, encoding='utf-8-sig')

print("Extraction termin√©e ! Donn√©es enregistr√©es dans 'catastrophes_nat_georisques_annuel.csv'")
```
